# Libraries

```python
import pandas as pd
import sqlite3 as sql3
import matplotlib.pyplot as plt
```

# Functions

```python
#GET THE DATA
def from_table_to_df(db_name, table_name):
    query = 'SELECT * FROM ' + table_name
    conn = sql3.connect(db_name)
    sql_query = pd.read_sql_query(query, conn)
    df = pd.DataFrame(sql_query) 
    return df
```

```python
weather_df = from_table_to_df('meteo.db', 'weather')
```

```python
#SAVE THE DATA ON CSV FILE
def from_df_to_csv(df, csv_path):
    df.to_csv(csv_path)
    print(f'Table stored as .csv at {csv_path}')
```

```python
from_df_to_csv(weather_df, 'weather.csv')
```

```python
#PLOT A BAR GRAPH
def plot_bars(df, var1, var1_name, var2, var2_name):
    
    plt.style.use('dark_background')
    plt.bar(var1, var2, 0.5)
    plt.xlabel(f"{var1_name}")
    plt.ylabel(f"{var2_name}")
    plt.title(f"{var1_name}-{var2_name} correlation")
    plt.xticks(rotation=90)
    
    plt.show()
```

```python
plot_bars(weather_df, weather_df["weather"], "Weather", weather_df["feels_like"], "Perceived temperature")
```
